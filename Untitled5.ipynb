```python
url='http://www.ibm.com/'
```

```python
s=requests.get(url)
```

```python
so=BeautifulSoup(s.text, 'html')
```

```python
print(so)
```

```python
print(so.prettify())
```

```python
p_tags=so.find_all('p')
```

```python
for tag in p_tags:
    print(tag)
```

```python
story_tags=so.find_all('p' ,attrs={'slot':'cta'})
for tag in story_tags:
    print(tag)

                       
                    
```
